```python
# Encontre duas formas de selecionar os dados #

# Forma 1 - Utilizando .loc filtrando por rótulo

dados.loc[dados['Bairro'] == 'Liberdade']
```

```python
# Forma 2 - Utilizando .iloc filtrando por index

dados.iloc[25:30]
```

```python
# Pegar outros estatísticas dos dados como mediana, mín, máx #

dados.describe()
```

```python
# Realizar a média da metragem para cada um dos bairros #

data_bairro = dados['Bairro'].values
dict_media_bairros = {}

for bairro in data_bairro:
  media_metragem = dados.loc[dados['Bairro'] == bairro, 'Metragem'].mean() # Calcula a metragem média para cada bairro
  dict_media_bairros[bairro] = media_metragem # Adiciona a média no dicionário

dict_media_bairros
```

```python
# Realizar a média da metragem para cada um dos bairros - Forma alternativa #

bairro_metragem = dados.groupby(['Bairro'])['Metragem'].agg('mean').round(2)
bairro_metragem
```

```python
# Descobrir quais bairros não tem nome de rua no DataFrame #

sem_rua = dados.loc[dados['Rua'].isnull(),'Bairro'].value_counts()
sem_rua
```

```python
# Explorar alguns gráficos na documentação e aplicar nas demais colunas do DataFrame (tentar colocar alguma conclusão) #

total_vagas = dados.groupby('Vagas').size()
total_vagas
```

```python
total_vagas.plot(kind='pie',figsize=(10,10),label='Total de Vagas')

# Chegamos a conclusão de que a maioria dos imóveis possuem até no máximo 4 vagas disponíveis
# Podemos supor que existem dados errados em algumas colunas pois encontramos imóveis com até 50 vagas disponíveis
```

```python
# Plot dos 10 bairros com mais ruas sem nome

bairros_mais_sem_rua = sem_rua.head(10)
bairros_mais_sem_rua.plot(kind="bar")
```

```python
# Plot dos 15 bairros com maiores metragens 

bairro_metragem.sort_values(ascending=False)
```

```python
bairro_metragem.sort_values(ascending=False).head(15).plot.bar();

# Podemos concluir que há um erro no Sítio Represa ou apenas 1 imóvel no local com metragem grande
```

# Aula 2 - Imersão de Dados (Alura)

```python
# Verificando os valores representados na coluna Valor dividindo suas partes

dados["Valor"].str.split(expand = True)
```

```python
# Verificar se na coluna 0 temos algum valor diferente de R$

dados["Valor"].str.split(expand = True)[0].unique()
```

```python
# Verificar o número de valores únicos na coluna 1

len(dados["Valor"].str.split(expand = True)[1].unique())
```

```python
dados["Valor"].str.split(expand = True)[2].unique()
```

```python
# Podemos criar novas colunas com os valores separados da coluna "Valor"

dados[["Moeda","Valor_anuncio","Tipo_anuncio"]] = dados["Valor"].str.split(expand = True)
dados.head()
```

```python
# Eliminaremos o anúncio cujo tipo não é "None"

dados[dados["Tipo_anuncio"].isnull()]
```

```python
# Provando que funcionou e só temos anúncios tipo None

dados[dados["Tipo_anuncio"].isnull()]["Tipo_anuncio"].unique()
```

```python
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
# Não conseguiremos fazer a média desses valores por ainda possuirem ponto e
# não serem do tipo float

# Precisamos antes remover esse ponto dos valores da tabela 

dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
```

```python
dados_vendas.info()
```

```python
# Plotando o gráfico para esses valores convertidos

dados_vendas["Valor_anuncio_float"].plot.hist(bins = 50)
```

```python
pd.set_option("display.precision", 2) # Reduzindo o número de casas decimais
pd.set_option('display.float_format', lambda x: '%.2f' % x) # Removendo a notação científica dos valores do tipo float

dados_vendas.describe()
```

```python
dados_vendas[["Rua", "Bairro", "Cidade"]].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_float", kde = True)
ax.set_title("Histograma Valor Imóvel")
plt.xlim((-50, 10000000))
plt.show()
```

```python
dados_vendas.head()
```

```python
# introduzindo a coluna de médias e pegando o valor por m² dos bairros de duas formas

dados_vendas["Valor_m2"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
dados_vendas.head()
```

```python
dados_vendas.groupby("Bairro").mean()
```

```python
dados_bairro = dados_vendas.groupby("Bairro").sum()
dados_bairro
```

```python
dados_bairro["Valor_m2_bairro"] = dados_bairro["Valor_anuncio_float"]/dados_bairro["Metragem"]
dados_bairro
```

```python
dados_vendas.groupby("Bairro").mean().loc["Vila Mariana"]
```

```python
dados_bairro.loc["Vila Mariana"]
```

```python
dados_bairro["Valor_m2_bairro"]["Vila Mariana"]
```

```python
# Plotando o m² dos 10 bairros mais frequentes

top_bairros = dados_vendas["Bairro"].value_counts()[:10].index
top_bairros
```

```python
# Os bairros nesse momento são index, logo precisamos resetar para voltar a ser uma coluna
# (inplace = true) substitui a necessidade de igualar o resultado na variável original

dados_bairro.reset_index(inplace = True)
dados_bairro.head()
```

```python
dados_bairro.query("Bairro in @top_bairros")
```

```python
plt.figure(figsize=(10, 8))
ax = sns.barplot(x="Bairro", y="Valor_m2_bairro", data=dados_bairro.query("Bairro in @top_bairros"))
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros"), x="Bairro", y="Valor_anuncio_float")
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas.query("Bairro in @top_bairros & Metragem < 30000"), x="Bairro", y="Metragem")
ax.tick_params(axis='x', rotation=45)
plt.show()
```
